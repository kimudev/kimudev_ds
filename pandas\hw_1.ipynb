## Задание 1
Датасет для домашнего задания вы найдете в разделе “Материалы к домашнему заданию” (файл “Дополнительные файлы для домашнего задания”) к этому занятию.
Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
ratings_of_movies = ratings.merge(movies, on='movieId', how='inner')

top_movie = ratings_of_movies[ratings_of_movies.rating == 5.0].groupby('movieId').count().sort_values('rating', ascending=False).merge(movies, on = 'movieId', how = 'inner')
top_movie[['movieId', 'title_y']].head(1)
```

## Задание 2
По данным файла power.csv (находится в архиве в разделе Материалы к лекции «Библиотека Pandas» - файл “Дополнительные файлы для домашнего задания”) посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
import pandas as pd

power = pd.read_csv('power.csv')

baltic_countries = power[(power.country.str.contains('Lit', case=True)) | (power.country.str.contains('Lat', case=True)) | (power.country.str.contains('Est', case=True))].country.unique()

sum_quantity = power[(power.country.isin(list(baltic_countries))) & (power.category.isin([4, 12, 21])) & (power.quantity >= 0) & (power.year.between(2005, 2010))].groupby('country').count().sum()
sum_quantity.quantity


```
